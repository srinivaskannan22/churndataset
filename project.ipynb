```python
import pandas as pd 

customer_churn_dataset=pd.read_csv("customer_churn_dataset-training-master.csv")
```

```python
customer_churn_dataset
```

```python
customer_churn_dataset=customer_churn_dataset.drop_duplicates(subset=["CustomerID"])
customer_churn_dataset=customer_churn_dataset.dropna()
```

```python
customer_churn_dataset=customer_churn_dataset.drop(columns="CustomerID")
```

```python
from sklearn.preprocessing import LabelEncoder
lb_gender=LabelEncoder()
lb_subsribtion=LabelEncoder()
lb_Contract=LabelEncoder()
```

```python
customer_churn_dataset["Gender"]=lb_gender.fit(customer_churn_dataset["Gender"])
customer_churn_dataset["Subscription Type"]=lb_subsribtion.fit(customer_churn_dataset["Subscription Type"])
customer_churn_dataset["Contract Length"]=lb_Contract.fit(customer_churn_dataset["Contract Length"])
customer_churn_dataset["Gender"]=lb_gender.transform(customer_churn_dataset["Gender"])
customer_churn_dataset["Subscription Type"]=lb_subsribtion.transform(customer_churn_dataset["Subscription Type"])
customer_churn_dataset["Contract Length"]=lb_Contract.transform(customer_churn_dataset["Contract Length"])
```

```python
from sklearn.preprocessing import MinMaxScaler
Ms=MinMaxScaler()
```
